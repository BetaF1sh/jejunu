## 라인 끝 부분에 `text(i)` 넣기

```java
// 미완성

void setup()
{
    size(300, 300);
    textSize(10);
    textAlign(CENTER);
}

void draw()
{
    background(255);
    translate(width/2, height/2);
    for (int i = 0; i < 360; i++) {
        line(0, 0, 0, -100);
        if (i % 90 == 0) {
            translate(0, -250);
            rotate(-radians(i));
            text(i, 0, 0);
        }
        rotate(radians(10));
    }
}
```
