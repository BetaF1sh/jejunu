```java
float[] cx, cy;
int mark;

void setup() {
    size(600, 600);
    cx = new float[20];
    cy = new float[20];
    for (int i = 0; i < 20; i += 1) {
        cx[i] = random(width);
        cy[i] = random(height);
    }
    textAlign(CENTER, CENTER);
    textSize(64);
    frameRate(1);
}

void draw() {
    for (int i = 0; i < 20; i += 1) {
        if (i == mark){
            fill(0, 255, 0);
        }else if (mark < 20){
            fill(255);
        }else{ mark = 0;}
//         if (boolean(i == mark)) ? fill(0, 255, 0) : fill(255);
        ellipse(cx[i], cy[i], 100, 100);
        fill(0);
        text(i, cx[i], cy[i]);
    }
    mark++;
}
```
