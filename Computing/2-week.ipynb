#### mouseClicked()
이벤트는 내장 함수로 따로 분류해두는게 바람직함

`mouseButton`, `width` 등은 예약된 내장 변수임.

```java
int cx;
int delta = 3;

void setup() {
  size(480, 280);
  cx = width / 2;
}

void draw() {
  background(255, 255, 0);
  cx += delta;
  ellipse(cx, height / 2, 100, 100);
  if (cx + 50 >= width || cx - 50 < 0) delta *= -1;
}

void mouseClicked() {
    if (mouseButton == LEFT) {
        noLoop();
    } else {
        loop();
    }
}
```

```java
int n;

void setup() {
    size(300, 300);
    textAlign(CENTER, CENTER);
    textSize(32);
    fill(0);
}

void draw() {
    background(255);
    text(n, width/2, height/2);
}

void mouseClicked() {
    if (mouseButton == LEFT) {
        n++;
    } else {
        n--;
    }
}
```
