## 포인터와 가까운 객체 찾아주는 `getNearest()` 만들어보기

가장 작은 `dist(x1, y1, x2, y2)` 의 i 주소값 색인하기
- https://processing.org/reference/dist_.html

`float minDist = width + height;`는 `size()` 로 정한 평면에서 존재할 수 없는 더미 값

```java
float[] cx, cy;
int nearest;

void setup() {
    size(600, 600);
    cx = new float[10];
    cy = new float[10];
    for (int i = 0; i < 10; i += 1) {
        cx[i] = random(50, width - 50); // 50 정도 여백을 줌.
        cy[i] = random(50, height - 50);
    }
}

void draw() {
    for (int i = 0; i < 20; i += 1) {
        if (i == nearest) //???
            fill(255, 0, 0);
        else
            fill(255);
        ellipse(cx[i], cy[i], 100, 100);
        fill(0);
    }
}

int getNearest()
{
    int min = 0;
    float minDist = width + height;
    for (int i = 0; i < 10; i++) {
        float d = dist(cx[i], cy[i], mouseX, mouseY);
        if (d < minDist) {
            minDist = d;
            min = i;
        }
    }
    return min;
}


void mousemoved()
{
    nearest = getNearest();
}
```
