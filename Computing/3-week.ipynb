```java
float cx, cy;

void setup()
{
    size(800, 500);
    cx = width/2;
    cy = height/2;
}

void draw()
{
    background(255);
    ellipse(cx, cy, 200, 200);
}

void mouseDragged()
{
    if (dist(cx, cy, mouseX, mouseY) <= 100) {
    cx += mouseX - pmouseX;
    cy += mouseY - pmouseY;
    }
}
```
