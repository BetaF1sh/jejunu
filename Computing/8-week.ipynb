# 굴러가는 원형 애니매이션

> p. 82 ~ 84, 융합 교육과 컴퓨팅의 이해

```java
// 원형
int bottom;
int r = 100;
int angle;
float cx, cy;

void setup()
{
    size(1000, 600);
    bottom = height / 2 + r;
    cx = 0;
    cy = bottom - r;
    angle = 0;
}

void draw()
{
    background(128);
    line(0, bottom, width, bottom);
    translate(cx, cy);
    rotate(radians(angle));
    ellipse(0, 0, 2*r, 2*r);
    line(0, 0, 0, -r);
    angle += 1;
    cx = r * radians(angle);
}
```

# 굴러가는 직사각형 애니매이션

```java
int boxW, boxH;
float orgX;
int bottom;
boolean horizontal;
int angle;

void setup()
{
    size(1000, 600);
    boxW = 300;
    boxH = 150;
    orgX = boxH;
    bottom = height / 2 + boxW / 2;
}

void draw()
{
    background(224);
    line(0, bottom, width, bottom);
    translate(orgX, bottom);
    rotate(radians(angle));
    if (horizontal) {
        rect(-boxW, -boxH, boxW, boxH);
    } else {
        rect(-boxH, -boxW, boxH, boxW);
    }
    if (angle == 90) {
        angle = 0;
        if (horizontal)
            orgX += boxH;
        else
            orgX += boxW;
        horizontal = !horizontal;
    } else angle += 1;
}
```

# 드레깅을 통한 원의 크기 변경

```java
float cx, cy, r, rwh;
boolean resizeable;

int delta = 3;

void setup()
{
    size(600, 600);
    cx = width/ 2;
    cy = height / 2;
    r = 200;
}

void draw()
{
    background(225);
    rwh = 2 * r;
    ellipse(cx, cy, rwh, rwh);
}

void mouseMoved()
{
    cursor(ARROW);
    resizeable = false;
    if (abs(dist(cx, cy, mouseX, mouseY) - r) <= delta) { // 이해해보기!
        cursor(HAND);
        resizeable = true;
    }
}

void mouseDragged()
{
    if (resizeable == false) return;
    r = dist(cx, cy, mouseX, mouseY);
}
```

# 바람개비 그리기

```java
color pcolor;

void setup()
{
    size(800, 600);
    pcolor = color(random(255), random(255), random(255));
}

void draw()
{
    drawPinwheel(200, 200, 150, 50, color(255, 0, 0)); 
    drawPinwheel(300, 450, 100, 30, color(0, 255, 0)); 
    drawPinwheel(500, 300, 250, 60, pcolor); 
}

void drawPinwheel(float x, float y, int w, int h, color c)
{
    fill(c);
    rect(x, y - w, h, w);
    rect(x, y, w, h);
    rect(x - h, y, h, w);
    rect(x - w, y - h, w, h);
}
```

# 회전하는 바람개비

```java
float[] px, py, pspeed, pangle;
int[] pwidth, pheight;
color[] pcolor;

void setup()
{
    size(800, 600);
    px = new float[5];
    py = new float[5];
    pwidth = new int[5];
    pheight = new int[5];
    pcolor = new color[5];
    pspeed = new float[5];
    pangle = new float[5];
    for (int i = 0; i < 5; i += 1) {
        pheight[i] = int(random(10, 50));
        pwidth[i] = int(random(2 * pheight[i], 5 * pheight[i]));
        px[i] = int(random(pwidth[i], width - pwidth[i]));
        py[i] = int(random(pwidth[i], height - pwidth[i]));
        pcolor[i] = color(random(255), random(255),
            random(255));
        pspeed[i] = random(0.5, 10);
    }
}

void draw()
{
    background(255);
    for (int i = 0; i < 5; i += 1) {
        pushMatrix();
        translate(px[i], py[i]);
        rotate(radians(pangle[i]));
        drawPinwheel(0, 0, pwidth[i], pheight[i], pcolor[i]); 
        popMatrix();
        pangle[i] += pspeed[i];
    }
}

void drawPinwheel(float x, float y, int w, int h, color c)
{
    fill(c);
    rect(x, y - w, h, w);
    rect(x, y, w, h);
    rect(x - h, y, h, w);
    rect(x - w, y - h, w, h);
}
```

# 사각형의 회전

```java
int theta;

void setup()
{
    size(600, 600);
    rectMode(CENTER);
}

void draw()
{
    background(224);
    translate(width/2, height/2);
    rotate(radians(theta));
    rect(0, 0, 200, 200);
    line(-100, -100, 100, 100);
    line(-100, 100, 100, -100);
    theta += 1;
}
```
